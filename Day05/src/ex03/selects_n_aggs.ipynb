```python
import pandas as pd
```

```python
df = pd.read_json('../../datasets/auto.json', orient = 'records')
df = df.set_index('CarNumber')
df
```

## Display

```python
df[df['Fines']>2100]
```

```python
df[(df['Fines'] > 2100) & (df['Refund'] == 2)]
```

```python
models = ['Focus','Corolla']
df[df['Model'].isin(models)]
```

```python
car_numbers = ['Y7689C197RUS', '92928M178RUS', '7788KT197RUS', 'H115YO163RUS', 'X758HY197RUS']
df[df.index.isin(car_numbers)]
```

### aggregations with the make and the model

### median fines by make

```python
df.groupby('Make')['Fines'].median()
```

### median fines by make and model

```python
df.groupby(['Make', 'Model'])['Fines'].median()
```

### number of fines by make and model

```python
df.groupby(['Make', 'Model'])['Fines'].count()
```

### min max fines by make and model

```python
df.groupby(['Make', 'Model'])['Fines'].agg(['min', 'max'])
```

```python
df.groupby(['Make', 'Model'])['Fines'].std()
```

## Aggregations with car number

### Car numbers by fines desc

```python
count = df.index.value_counts().sort_values(ascending=False)
count
```

### Top 1 count

```python
df[df.index == count.index[0]]
```

### Total fines

```python
fines = df.groupby(df.index)['Fines'].sum().sort_values(ascending=False)
fines
```

### Top 1 fine

```python
df[df.index == fines.index[0]]
```

### Unique check

```python
counts = df.groupby(df.index)['Model'].nunique() >1
counts[counts>1]
```
