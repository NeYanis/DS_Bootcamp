```python
import pandas as pd
```

```python
df = pd.read_csv('../../datasets/auto.csv', index_col = 'ID', sep = ',')
df.head()
```

```python
df.count()
```

### cleaned

```python
df = df.drop_duplicates(subset=['CarNumber', 'Make_n_model', 'Fines'], keep = 'last')
df.count()
```

### Work with missing

```python
df.isnull().sum()
```

```python
df = df.dropna(axis=1, thresh= len(df)-500)
df.isnull().sum()
```

```python
df['Refund'] = df['Refund'].ffill()
mean = df['Fines'].mean()
df['Fines'] = df['Fines'].fillna(mean)
df.isnull().sum()
```

```python
df[['Make','Model']] = df['Make_n_model'].apply(lambda x: pd.Series(str(x).split(' ',1)))
df = df.drop('Make_n_model', axis=1)
df.head()
```

```python
df.to_json('../../datasets/auto.json', orient = 'records')
```
