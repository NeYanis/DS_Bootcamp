````python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score,roc_auc_score
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, VotingClassifier, BaggingClassifier, StackingClassifier
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.linear_model import LogisticRegression
import joblib
````

## 1. Preprocessing

1. Create the same dataframe as in the previous exercise.
2. Using `train_valid_split` with parameters `valid_size=0.2`, `random_state=21` get `X_train`, `y_train`, `X_valid`, `y_valid` and then get `X_train`, `y_train`, `X_valid`, `y_valid` from the previous `X_train`, `y_train`. Use the additional parameter `stratify`.

````python
df = pd.read_csv('../../datasets/day-of-week-not-scaled.csv')
day = pd.read_csv('../../datasets/dayofweek.csv')
df['dayofweek'] = day['dayofweek']
df
````

````python
X = df.drop('dayofweek',axis=1)
y = df['dayofweek']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, stratify=y)
````

````python
XX=X_train.copy()
yy=y_train.copy()
X_train, X_valid, y_train, y_valid = train_test_split(XX,yy,test_size=0.2, stratify=yy)
````

## 2. Individual classifiers

````python
svc = SVC(C=10,class_weight=None,gamma='auto', kernel='rbf', probability=True, random_state=21)
svc.fit(X_train,y_train)
y_pred = svc.predict(X_valid)
print(f'accuracy is {accuracy_score(y_valid, y_pred):.5f}')
print(f'precision is {precision_score(y_valid, y_pred, average='weighted'):.5f}')
print(f'recall is {recall_score(y_valid, y_pred, average='weighted'):.5f}')
````

````python
dt = DecisionTreeClassifier(class_weight='balanced',criterion='gini', max_depth=22, random_state=21)
dt.fit(X_train,y_train)
y_pred = dt.predict(X_valid)
print(f'accuracy is {accuracy_score(y_valid, y_pred):.5f}')
print(f'precision is {precision_score(y_valid, y_pred, average='weighted'):.5f}')
print(f'recall is {recall_score(y_valid, y_pred, average='weighted'):.5f}')
````

````python
rf = RandomForestClassifier(class_weight=None,criterion='gini', max_depth=28,n_estimators=50, random_state=21)
rf.fit(X_train,y_train)
y_pred = rf.predict(X_valid)
print(f'accuracy is {accuracy_score(y_valid, y_pred):.5f}')
print(f'precision is {precision_score(y_valid, y_pred, average='weighted'):.5f}')
print(f'recall is {recall_score(y_valid, y_pred, average='weighted'):.5f}')
````

## 3. Voting classifiers

1. Using `VotingClassifier` and the three models that you have just trained, calculate the `accuracy`, `precision`, and `recall` on the validation set.
2. Play with the other parameteres.
3. Calculate the `accuracy`, `precision` and `recall` on the test set for the model with the best weights in terms of accuracy (if there are several of them with equal values, choose the one with the higher precision).

````python
vot = VotingClassifier(estimators=[('svm',svc),('des_tree',dt),('r_forest',rf),], voting='hard')
````

````python
vot.fit(X_train,y_train)
y_pred = vot.predict(X_valid)
print(f'accuracy is {accuracy_score(y_valid, y_pred):.5f}')
print(f'precision is {precision_score(y_valid, y_pred, average='weighted'):.5f}')
print(f'recall is {recall_score(y_valid, y_pred, average='weighted'):.5f}')
````

````python
weight = [(3,1,2),(1,2,1),(1,2,2),(2,2,1),(4,1,1),(4,1,4)]
best_ac = 0
best_pr = 0
best_weight = None
for i in weight:
    wvot = VotingClassifier(estimators=[('svm',svc),('des_tree',dt),('r_forest',rf)],voting='soft',weights=i)
    wvot.fit(X_train,y_train)
    pred = wvot.predict(X_valid)
    ac = accuracy_score(y_valid,pred)
    pr =precision_score(y_valid, pred, average='macro')
    if ac > best_ac or(ac == best_ac and pr > best_pr):
        best_ac = ac
        best_pr = pr
        best_weight = i
print(f'accuracy is {best_ac:.5f}')
print(f'precision is {best_pr:.5f}')
print(f'weight is {best_weight}')
````

````python
vot = VotingClassifier(estimators=[('svm',svc),('des_tree',dt),('r_forest',rf),],voting='soft',weights=best_weight)
vot.fit(X_train,y_train)
y_pred = vot.predict(X_test)
print(f'accuracy is {accuracy_score(y_test, y_pred):.5f}')
print(f'precision is {precision_score(y_test, y_pred, average='macro'):.5f}')
print(f'recall is {recall_score(y_test, y_pred, average='macro'):.5f}')
````

## 4. Bagging classifiers

3. Calculate the `accuracy`, `precision`, and `recall` for the model with the best parameters (in terms of accuracy) on the test set (if there are several of them with equal values, choose the one with the higher precision)

````python
svc = SVC(C=10,class_weight=None,gamma='auto', kernel='rbf', probability=True, random_state=21)
bc = BaggingClassifier(estimator=svc, n_estimators=10, random_state=21)
bc.fit(X_train,y_train)
y_pred = bc.predict(X_test)
print(f'accuracy is {accuracy_score(y_test, y_pred):.5f}')
print(f'precision is {precision_score(y_test, y_pred, average='weighted'):.5f}')
print(f'recall is {recall_score(y_test, y_pred, average='weighted'):.5f}')
````

````python
params = {'n_estimators': range(2,50),
          'bootstrap': [True, False]}
gs = GridSearchCV(estimator=bc,param_grid=params,scoring='accuracy', verbose=2)
gs.fit(X_train,y_train)
print(f'best score {gs.best_score_}')
print(f'parameters {gs.best_params_}')
````

## 5. Stacking classifiers

1. To achieve reproducibility in this case you will have to create an object of cross-validation generator: `StratifiedKFold(n_splits=n, shuffle=True, random_state=21)`, where `n` you will try to optimize (the details are below).
2. Using `StackingClassifier` and the three models that you have recently trained, calculate the `accuracy`, `precision` and `recall` on the validation set, try different values of `n_splits` `[2, 3, 4, 5, 6, 7]` in the cross-validation generator and parameter `passthrough` in the classifier itself,
3. Calculate the `accuracy`, `precision`, and `recall` for the model with the best parameters (in terms of accuracy) on the test set (if there are several of them with equal values, choose the one with the higher precision). Use `final_estimator=LogisticRegression(solver='liblinear')`.

````python
res=[]
for n in range(2,8):
    for passthrough in [True, False]:
        kf = StratifiedKFold(n_splits=n, shuffle=True, random_state=21)
        stack = StackingClassifier(estimators=[('svm',svc),('des_tree',dt),('r_forest',rf)],final_estimator=LogisticRegression(solver='liblinear'), cv = kf, passthrough=passthrough)
        stack.fit(X_train,y_train)
        y_pred = stack.predict(X_valid)
        ac = accuracy_score(y_valid,pred)
        pr =precision_score(y_valid, pred, average='macro')
        rec = recall_score(y_valid,y_pred,average='macro')
        res.append((n,passthrough,ac,pr,rec))
````

````python
best = max(res,  key= lambda x: (x[2], x[3]))
print(f'Best: n_splits: {best[0]}, passtrough: {best[1]}\naccuracy: {best[2]}, precision: {best[3]}, recall: {best[4]}')
````

````python
kf = StratifiedKFold(n_splits=best[0], shuffle=True, random_state=21)
stack = StackingClassifier(estimators=[('svm',svc),('des_tree',dt),('r_forest',rf)],final_estimator=LogisticRegression(solver='liblinear'), cv = kf, passthrough=best[1])
stack.fit(pd.concat([X_train, X_valid]), pd.concat([y_train, y_valid]))
y_pred = stack.predict(X_test)
print(f'accuracy is {accuracy_score(y_test, y_pred):.5f}')
print(f'precision is {precision_score(y_test, y_pred, average='macro'):.5f}')
print(f'recall is {recall_score(y_test, y_pred, average='macro'):.5f}')
````
